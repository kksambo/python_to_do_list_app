```python
#importing gradio and giving it a simple name gr
import gradio as gr
```

```python

'''
this class is a blueprint for making tasks
every takes created must have title, date and duration


'''

class Task:

  '''
  this method will be called everytime a new task is created
  is used for constructing objects of the Task class
  '''
  def __init__(self, title, date, duration,status):
    self.title = title
    self.date = date
    self.duration = duration
    self.status = status

  '''
  the edit method allows user to edit details of a specific task
  it receives three parameters title string, date string and duration int

  '''

  def edit(self, title , date , duration,status):
    self.title = title
    self.date = date
    self.duration = duration
    self.status = status


  '''
  this method is used o properly format the task details

  '''

  def __str__(self):
    return f"Title: {self.title}\nDate: {self.date}\nDuration: {self.duration} \nStatus: {self.status}"
```

```python

#samples tasks
task1 = Task("cooking", "01/01/2000", 4,"to_do")
task2 = Task("baking", "01/01/2000", 4,"to_do")
task3 = Task("washing", "01/01/2000", 4,"to_do")
task4 = Task("shopping", "01/01/2000", 4,"to_do")
tasks = [task1, task2, task3, task4]


'''
the task_to_string method format the tasks as one string with task id
it receives one parameter tasks list
'''
def task_to_string(tasks):
  task_string = ""
  task_id = 1
  for task in tasks:
    task_string += "task_id : "+str(task_id)+"\n" + str(task) + "\n=================================\n"
    task_id += 1
  return task_string


'''
the edit_input_sanitize method gives the user freedom to update task attributes
without affecting other attributes
'''
def edit_input_sanitize(task :Task,title,date,duration,status):
  if title == "":
    title = task.title
  if date == "":
    date = task.date
  if duration == 0:
    duration = task.duration

  task.edit(title,date,duration,status)



'''
the task_manager method gives user the power to add,edit,view and delete tasks
it receives six parameters task_id string,title string,date string,duration int,status string,option string
no need for the user to insert all parameters especially when edititng deleting or viewing
'''
def task_manager(task_id,title,date,duration,status,option):
  if option == "edit":
    if task_id == "":
      return "enter task id"
    else:
      task_id = int(task_id)#changing task_id to an int so that i can use it as index

      task = tasks[task_id-1]# m subtracting 1 here because indices starts from 0
      edit_input_sanitize(task,title,date,duration,status)

      return task_to_string(tasks)


  elif option == "add":

    task = Task(title,date,duration,status)
    tasks.append(task)
    return task_to_string(tasks)

  elif option == "delete":
    if task_id == "":
      return "enter task id"
    else:
      task_id = int(task_id)
      tasks.pop(task_id-1)
      return task_to_string(tasks)

  else:
    return task_to_string(tasks)

'''
the interface object holds and instance of gradio interface
helps us to give user a simple interface to the application and also provide
some abstruction users only see what is useful to them
'''
interface = gr.Interface(
    fn=task_manager,
    inputs=["text","text", "text", "number", gr.Dropdown(label="status", choices=["to do", "in progress","completed"], value="to do")
    ,gr.Dropdown(label="option", choices=["add", "edit","delete"], value="view all")],
    outputs="text")

interface.launch()
```
